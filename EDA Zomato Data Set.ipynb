```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country, on='Country Code', how='left')
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
final_df.Country.value_counts().index
```

```python
country_names = final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().values
```

```python
country_values = final_df.Country.value_counts().values
```

```python
#pie chart
plt.pie(country_values,labels=country_names)
```

```python
#pie chart top 3
plt.pie(country_values[:3],labels=country_names[:3])
```

```python
#pie chart top 3
plt.pie(country_values[:3],labels=country_names[:3], autopct='%1.2f%%')
```

Observation: Zomato maximum transactions are from India, followed by USA and India

```python
ratings = final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Maximum number of people have not given any ratings<br>
whenever the raitng is between 4.5 to 4.9 it's excellent<br>
when rating is between 4.0 to 3.4, it's very good

```python
ratings.head()
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color', data=ratings, palette=['white','red','orange','yellow','green','green'])
    
```

Observation:
1. Not rated is very high
2. Maximum number of rating are between 2.5 to 3.4

```python
# Count plot, frequency in the data frame and not the overal count
sns.countplot(x='Rating color',data=ratings, palette=['blue','red','orange','yellow','green','green'])
```
